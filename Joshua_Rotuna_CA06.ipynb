```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import NearestNeighbors
```

### Reading in Dataset

```python
# Reading in Dataset and examining first few rows
df = pd.read_csv('https://github.com/ArinB/CA05-kNN/raw/master/movies_recommendation_data.csv')
df
```

```python
# Dataset Shape and Info
print(df.shape)
print()
print(df.info())
```

### Creating Feature Matrices

```python
  # Defining Features Matrix:
features_matrix = df.iloc[:,2:-1]
  
  # Creating Test Matrix for Prediction:
test_matrix = [7.2, 1,1,0,0,0,0,1]
test_matrix = np.reshape(test_matrix, (-1, 1)).T
test_matrix
```

## Building KNN Model

```python
# Importing KNN from Scikit-learn
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5, algorithm='brute')  # Using Brute force method because it is a very small dataset
knn.fit(features_matrix, df['Movie Name'])
pred_values = knn.predict(test_matrix)


# Nearest neighbor:
from sklearn.neighbors import NearestNeighbors
knn = NearestNeighbors(n_neighbors=5)     # n_neighbors adjusts the amount of "recommendations" it will make
knn.fit(features_matrix)        # Fit the model with all necessary features
neighbors = knn.kneighbors([[7.2, 1,1,0,0,0,0,1]])    # Creading prediction from hypothetical movie's feature list
neighbors = neighbors[1]    #GRabbing only the array with column indexes

#Looping through list of indexes and returning the Movie Name:
for x in neighbors:
  y = df.iloc[x,1]
  movie_rec_df = pd.DataFrame(y)
  print("5 Movies you may like as well:")
  print(movie_rec_df)

```
